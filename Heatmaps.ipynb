```python
import numpy as np
import numpy.random
import matplotlib.pyplot as plt

def showHeatmap(df,xnp,ynp,objname):

  objdf = df[df['objname'] == objname]
  xnp = objdf['gaze_coordx'].to_numpy()
  ynp = objdf['gaze_coordy'].to_numpy()

  xs = np.array(xnp, dtype=float)
  ys = np.array(ynp, dtype=float)

  minx = objdf['minboundx'].to_numpy()
  minx = np.array(minx , dtype=float)
  maxx = objdf['maxboundx'].to_numpy()
  maxx = np.array(maxx , dtype=float)
  miny = objdf['minboundy'].to_numpy()
  miny = np.array(miny , dtype=float)
  maxy = objdf['maxboundy'].to_numpy()
  maxy = np.array(maxy , dtype=float)

  heatmap, xedges, yedges = np.histogram2d(xs, ys, bins=6)
  #extent = [np.min(xs), np.max(xs), np.min(ys), np.max(ys)]
  extent = [np.min(minx), np.max(maxx), np.min(miny), np.max(maxy)]

  plt.clf()
  plt.title(objname + " "+ " Eye gaze")
  plt.imshow(heatmap.T, extent=extent, origin='lower')
  plt.show()
 
```

```python
import pandas as pd

df = pd.read_csv("eyeDataDoc.csv")
df['gaze_coordx'] = df['gaze_coordx'].str[1:]
df['gaze_coordz'] = df['gaze_coordz'].str[:-1]
df['minboundx'] = df['minboundx'].str[1:]
df['minboundz'] = df['minboundz'].str[:-1]
df['maxboundx'] = df['maxboundx'].str[1:]
df['maxboundz'] = df['maxboundz'].str[:-1]

xnp = df['gaze_coordx'].to_numpy()
ynp = df['gaze_coordz'].to_numpy()
for name in df['objname'].unique():
  showHeatmap(df,xnp,ynp,name)

df.head(5)
```
